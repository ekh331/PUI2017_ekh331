```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D

%pylab inline
```

```python
df2 = pd.read_csv('https://data.cityofchicago.org/api/views/w8km-9pzd/rows.csv?accessType=DOWNLOAD')
```

```python
df2.head(2)
```

```python
plt.figure(figsize=(12,8))

total, = plt.plot(df2.year, df2.total, label='Total', linewidth=4,)
bus, = plt.plot(df2.year, df2.bus, label='Bus', linewidth=4)
rail, = plt.plot(df2.year, df2.rail, label='Train', linewidth=4)

labels = range(1, 8)
total.axes.yaxis.set_ticklabels(labels)
bus.axes.yaxis.set_ticklabels(labels)
rail.axes.yaxis.set_ticklabels(labels)

plt.legend(prop={'size': 15})
plt.xlabel('Year', size=18)
plt.ylabel('Annual Ridership (millions)', size=18)
plt.title('Chicago Public Transit Annual Ridership, 1988-2016', size=22)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(ymax=7e8)

ax = plt.gca()
ax.grid(color='gray', linestyle='-', linewidth=1, alpha=0.3)

plt.show()

```

#### Figure 1: Chicago Transit Authority (CTA) annual ridership estimates by type, 1988-2016. While overall ridership has decreased over this time period, train ridership across Chicago's elevated ('L') lines has increased in tandem with a decrease in bus ridership.
